```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Shubham Vartak (Friend).txt','r',encoding= 'utf-8')
```

```python
data = f.read()
```

```python
print(data) 
```

```python
pattern= '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern,data)[1:]
```

```python
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_message':messages,'message_date':dates})
#df has two columns userMessage and messageDate
```

```python
df['message_date']
```

```python
df['message_date'] = pd.to_datetime(df['message_date'],format = '%m/%d/%y, %H:%M - ')
```

```python
df
```

```python
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour + 1))
    else:
        period.append(str(hour) + "-" + str(hour + 1))

df['period'] = period
```

```python
def emoji_helper(selected_user,df):
    if selected_user != 'Overall':
        df = df[df['user'] == selected_user]

    emojis = []
    # for message in df['message']:
    #    emojis.extend([c for c in message if c in emoji.UNICODE_EMOJI['en']])
    for message in df['message']:
        emojis1 = emoji.distinct_emoji_list(message)
        emojis.extend([is_emoji for is_emoji in emojis1])
    print(emojis)
    emoji_df = pd.DataFrame(Counter(emojis).most_common(len(Counter(emojis))))

    return emoji_df
```
